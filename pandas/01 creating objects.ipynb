### object creation

#### pd.series()

```python
# Series: 1d labeled array - heterogeneous - default index: integer starting with 0
s = pd.Series([1, 3, 5, np.nan, 6, 8])
s
```

```python
# to change index
s = pd.Series(data = [10,20,30,40],
              index = ['a','b','c','d'])
s
```

```python
# s = pd.Series(data = [10,20],
#               index = ['a','b','c','d'])
# # ValueError: Length of values (2) does not match length of index (4)
```

```python
# specify values for each column
emp_dict = {"id": [1051, 1052, 1053],
           "name": ['tony', 'john', 'alen'],
           "age": [30, 25, 34],
           "salary": [45000, 38000, 60000]}
df = pd.DataFrame(data = emp_dict,
                  index= [1, 2, 3]) # dict values length
df
```

#### pd.DataFrame()

```python
#### pd.DataFrame()# specify values for each row
df = pd.DataFrame(data = [[1051, 'tony', 30, 45000],
                          [1052, 'john', 25, 38000],
                          [1053, 'alen', 34, 60000]],
                  index= ['1st employee', '2nd employee', '3rd employee'],
                  columns=['id', 'name', 'age', 'salary'])
df
```

#### pd.DataFrame() - multiindex

```python
# data frame with multi-index
df_multiindex = pd.DataFrame(data = emp_dict,
                  index = pd.MultiIndex.from_tuples(
                            [('NorthIndia',101),('NorthIndia',102),("SouthIndia",1001)],
                            names = ['Region','Subregion_code']
                          )
                  )
df_multiindex
```
