```python
import pandas as pd
```

```python
rectangle = {
    'length': [2, 4, 5, 7, 3, 6, 9, 8, 1, 5],
    'width': [3, 5, 2, 6, 4, 8, 1, 9, 7, 2]
}

df = pd.DataFrame(rectangle)
df
```

### df.assign(newcol = lambda, newcol2 = 'constvalue')

```python
# make 1 or more new columns

df = df.assign(area = lambda df: df.length * df.width, 
               lenSquare = lambda df: df.length * df.length, 
               name = 'rectangle')
df
```

### df['newcolumn'] = df.oldcol * df.oldcol2

```python
# add single column
df['sqWidth'] = df.width * df.width
df
```

### df.newcolumn = pd.qcut(df.col, q=n, labels=False)

```python
# bin column into n bucketsn

n=4
df['Bins'] = pd.qcut(df.area, q=2, labels = False)
```
