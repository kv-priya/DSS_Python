```python
import pandas as pd
import numpy as np
```

```python
# Create the DataFrame
data = {
    'Name': ['John', 'Emily', 'David', 'Sophia', 'Michael', 'Emma', 'Daniel', 'Olivia', 'James', 'Ava'],
    'Age': [25, 32, 45, 28, 36, 19, 52, 40, 33, 27],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'Country': ['USA', 'Canada', 'UK', 'Australia', 'Germany', 'France', 'India', 'Brazil', 'Japan', 'Mexico'],
    'Salary': [50000, 60000, 70000, 55000, 65000, 45000, 80000, 55000, 60000, 50000],
    'Department': ['IT', 'Sales', 'HR', 'IT', 'Finance', 'Marketing', 'IT', 'HR', 'Finance', 'Sales']
}

df = pd.DataFrame(data)
df
```

### df.reindex(list_of_new_index)

```python
# Reindexing is the process of changing the row or column labels of a DataFrame

new_index = [8, 3, 2, 0, 4, 5, 9, 1, 7, 6] # ids
df_reindexed = df.reindex(new_index)
df_reindexed
```

```python
df # reindex doesnot affect original dataframe
```

### df.reset_index()

```python
# reset index

df_resetindex = df_reindexed.reset_index() # sets row numbers as index moving current index to columns
df_resetindex 
```

```python
df_reindexed # reindexed doesnot affect original dataframe
```

### df.sort_values(by='col',ascending=True)

```python
# Sorting is the process of arranging the rows or columns of a DataFrame in a specific order.
# based on 1 or more columns

# Sort the DataFrame by 'Age'column in ascending order
df_sorted = df.sort_values(by='Salary')
df_sorted
```

### df.sort_values(by=[cols],ascending=[booleans_for_each_col])

```python
# sort by multiple columns

# Sort the DataFrame by 'Age' in ascending order and 'Salary' in descending order
df_sorted = df.sort_values(by=['Salary', 'Age'], ascending=[False, True])
df_sorted
```

### df.sort_index(ascending=False)

```python
df_sortindex_desc = df.sort_index(ascending=False)
df_sortindex_desc
```

### df.rename(columns={'oldcol':'newcol'})

```python
# Renaming allows you to change the labels of rows or columns in a DataFrame.

# Rename the 'Salary' column to 'Annual Salary'
df_renamed = df.rename(columns={'Salary': 'Annual Salary', 'Department':'Department Name'})
df_renamed
```
