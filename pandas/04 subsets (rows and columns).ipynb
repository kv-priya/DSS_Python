```python
import numpy as np
import pandas as pd
```

**df.loc[] and df.iloc[] - ro select only rows or only columns or both**

**df.at[] and df.iat[] - to access a single value by row and column**

```python
### subset observations (rows)# specify values for each column
import random
import string

emp_dict = {"id": ['EBDA'+str(x) for x in range(1001,1031)],
           "name": [random.choice(string.ascii_letters.lower())*3 for x in range(30)],
           "age": [random.randint(22,45) for _ in range(30)],
           "salary": [random.randint(25000,90000) for _ in range(30)]}

df = pd.DataFrame(data = emp_dict,
                  index= list(range(1,31))) # dict values length
df.head()
```

### df.loc[rowindex:rowindex, colname:colname]

### df.iloc[rowindex:rowindex,colindex:colindex]

```python
# specified rows and columns
df.loc[1:5,'id':'age']
# df.iloc[0:5,0:3] # index starts with 0 and stop index in slice is exclusive
```

```python
# all rows and specified columns
df.loc[:,'id':'age']
df.iloc[:,0:3]
```

```python
# all rows and specified columns
df.loc[:,['id','age']]
# df.iloc[:,[0,2]]
```

### df.loc[rowindex:rowindex] - all columns

### df.iloc[rowindex:rowindex] - all columns

```python
# specified rows and all columns
df.loc[1:5]
df.iloc[0:5]
```

### df.iloc[[],:] - specified rows and all columns

```python
# specified rows and all columns
df.loc[[1,2,4],:]
df.iloc[[0,1,3],:]
```

### df.loc[rowcond, [columns]]

```python
# select rows meeting the condition, and specified columns
df.loc[df.age<30, ['id','age']]
```

```python
# select rows meeting the condition, and all columns
# df.loc[df.age<30, :] # this is same as below code
df[df.iloc[:,2] < 30]
```

### df.loc[rowcond, df.columns[:]]

```python
# using loc and index
df.loc[df.age<30, df.columns[0:3]]
```

### df.loc[].iloc[]

```python
# using loc and iloc
df.loc[:,'id':'age'].iloc[11:16,1:2]
```

### df.at[label,column] - access single value

```python
# access single value by label
df.at[3,'id']
```

### df.iat[rowindex,columnindex] - access single value

```python
# access single value by index
df.iat[2,0]
```
