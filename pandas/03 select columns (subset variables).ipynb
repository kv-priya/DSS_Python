### subset attributes (columns)

```python
import numpy as np
import pandas as pd
```

```python
### subset observations (rows)# specify values for each column
import random
import string

emp_dict = {"id": ['EBDA'+str(x) for x in range(1001,1031)],
           "name": [random.choice(string.ascii_letters.lower())*3 for x in range(30)],
           "age": [random.randint(22,45) for _ in range(30)],
           "salary": [random.randint(25000,90000) for _ in range(30)]}

df = pd.DataFrame(data = emp_dict,
                  index= list(range(1,31))) # dict values length
```

```python
df
```

### df[['singlecolumn']]

```python
# select single column with specific name
# df.age.head()
# df['age'].head()
df[['age']].head()
```

### df[['col1','col2']]

```python
# select multiple columns with specific names
df[['salary','age', 'id']].head()
```

#### df.filter(items=['c1','c2'])

```python
# select columns whose name matches regular expression regex
df.filter(items=['salary','age']).head() # select col by name
```

### df.filter(regex='',axis=1)

```python
df.filter(regex='^a').head() # select col by regex
# df.filter(regex='^a.*', axis=1).head() # select col by regex
# ^ starts with
# . => any 1 character
# * => 0 or more characters
```

### df.filter(like='',axis=1)

```python
df.filter(like='sal',axis=1).head() # select column by regex to index
```
