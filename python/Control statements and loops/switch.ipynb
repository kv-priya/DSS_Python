#### switch

```python
def http_error(status):
  match status: # run in colab
        case 400:
            print("case 400 block")
            return "Bad request"
        case 404:
            return "Not found"
        case 418:
            return "I'm a teapot"
        case _:
            return "Something's wrong with the internet"
```

```python
print(http_error(450))
# 200 status code: success
# 400 status code: bad request (url check)
# 500 status code: server side error
```
