#### string slicing

```python
# works on group of characters in the string
# a[startindex:endindex:step] {startindex : inclusive, endindex : exclusive}=> a[0:10:2] => 0,2,4,6,8
# range(0,6) => 0,1,2,3,4,5 (6 is exclusive, 1 is inclusive)


b = "Hello, World!"
print(b[2:])

b = "0123456789"
print(b[2:5]) # start is inclusive, end is exclusive
print(b[2:8:2]) # step 2

b = "987654321"
# negtive index slicing with same step direction
print(b[-1:-5:-2] , # -1, -1-2 = -3, -3-2 = -5 (exclusive - donot consider)
      b[-5:-1] , # -5, -5+1=-4, ...., -2
      b[:], # full string (start = 0, end = len-1, step = 1)
      b[::-1]) # -1 reverse direction (start = -1, end=len())

print(b[-5:-1:-1], # "" diff direction
      b[2:8:-5], # ""
      b[6:-2:-1]) # ""
```

#### maketrans()
- returns a mapping table that can be used with the translate() method to replace specified characters.
- Syntax : maketrans(str1, str2, str3)
- replace list of characters in str1 with list of characters in str2 and delete str3 list of characters

```python
# Create a mapping table, and use it in the translate() method
# to replace any "S" characters with a "P" character:
txt = "Hello Sam!"
mytable = txt.maketrans("S", "P")
print(txt.translate(mytable))
```

```python
# Use a mapping table to replace many characters:
txt = "Hi S-a-m!"
x = "mSa"
y = "eJo"
mytable = txt.maketrans(x, y)
print(txt.translate(mytable))
print(txt)
```

```python
# The third parameter in the mapping table describes characters that you want to remove from the string:
txt = "Good night Sam!"
x = "mSaG" # old
y = "eJoH" # new
z = "odnght" # to delete
mytable = txt.maketrans(x, y, z)
print(txt.translate(mytable))
print(txt)

# txt: Good night Sam!
# replace: Hood night Joe!
# delete: H i Joe
```

```python
# The maketrans() method itself returns a dictionary describing each replacement, in unicode:
txt = "Good night Sam!"
x = "mSa"
y = "eJo"
z = "odnght"
print(txt.maketrans(x, y, z)) # dictionary: mapping table
```

```python
txt = "banana,,,,,ssqqqww....."
x = txt.rstrip(",.qsw")
print(x)

txt = ",,,,,ssaOaww.....bananassssssss" # assssssss is edge
x = txt.rstrip(",.asw")
print(x)
```

#### center()
- Returns a centered string
- string.center(length, character)

```python
# Print the word "banana", taking up the space of 20 characters,
#  with "banana" in the middle
# string.center(length, character)
txt = "banana"
x = txt.center(20)
print(x)
```

```python
txt = "banana"
x = txt.center(21, "*")
print(x)
```

#### ljust()
- will left align the string, using a specified character (space is default) as the fill character.
- string.ljust(length, character)

```python
txt = "banana"
x = txt.ljust(20)

print(x + "is my favorite fruit.")
```

```python
txt = "banana"
x = txt.ljust(20, "*")
print(x)
```

```python
txt = "banana"
x = txt.ljust(3, "*") # no data loss
print(x)
```

#### rjust()
- will right align the string, using a specified character (space is default) as the fill character

```python
txt = "banana"
x = txt.rjust(20)

print(x, "is my favorite fruit.")
```

```python
txt = "banana"
x = txt.rjust(20, "*")

print(x)
```

#### zfill()
- Fills the string with a specified number of 0 values at the beginning
- string.zfill(len)

```python
txt = "24"
x = txt.zfill(1)
print(x)

x = txt.zfill(4) # output has minimum 4 characters
print(x)

# 30
# 150
# 35000

# 00030
# 00150
# 35000
```

```python
a = "hello"
b = "welcome to the jungle"
c = "10.000"

print(a.zfill(10))
print(b.zfill(10)) # output has minimum 10 characters
print(c.zfill(10))
```

```python
print("999".zfill(5))
print("4".zfill(5))
```
