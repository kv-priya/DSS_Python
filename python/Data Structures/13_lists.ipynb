### List

- List items are ordered, changeable (mutable), and allow duplicate values.
- List items are indexed, the first item has index [0], the second item has index [1] etc.
- Lists are ordered. The new items will be placed at the end of the list.
- Mutable. Can add, and remove items in a list after it has been created.
- Allows duplicates

```python
thislist = ["apple", "banana", "cherry", "apple"]
print(thislist, type(thislist))
```

#### slicing

- read range of list of items

```python
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[2:5]) # 2,3,4

#This will return the items from position 2 to 5.

#Remember that the first item is position 0,
#and note that the item in position 5 is NOT included
```

```python
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[:4]) # 0,1,2,3

#This will return the items from index 0 to index 4.

#Remember that index 0 is the first item, and index 4 is the fifth item
#Remember that the item in index 4 is NOT included
```

```python
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[2:]) # 2,3,4,5,....., len(thislist)-1

#This will return the items from index 2 to the end.

#Remember that index 0 is the first item, and index 2 is the third
```

```python
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[-4:-1]) # [-4:-1:1] => -4, -4+1=-3, -3+1=-2

#Negative indexing means starting from the end of the list.

#This example returns the items from index -4 (included) to index -1 (excluded)

#Remember that the last item has the index -1,
```

#### extend()
- Add the elements of a list (or any iterable), to the end of the current list

```python
list1 = ["a", "b" , "c"]
list2 = [1, 2, 3]

list1.extend(list2)
print("Extend:", list1)

list1 = ["a", "b" , "c"]
list1.append(list2)
print("Append:",list1)
```

```python
thislist = ["apple", "banana", "cherry", 'papaya']
tropical = ["mango", "pineapple", "papaya"]
thislist.extend(tropical) # add list of items from other list to the end
print(thislist)
```

#### copy()
- Returns a copy of the list

```python
l1 = [1,2,3]
l2 = l1               # l2 refers to same l1 mem loc
print(id(l1),id(l2))

l1 = [1,2,3]
l2 = l1.copy()        # l2 refers to diff mem loc where copy of l1 is created
print(id(l1),id(l2))
```

```python
thislist = ["apple", "banana", "cherry"]
mylist = thislist.copy()
print(mylist)
print(id(thislist) == id(mylist)) # False
```

```python
thislist = ["apple", "banana", "cherry"]
mylist = list(thislist) # list() creates new list with specified values
print(mylist)
print(id(thislist) == id(mylist))
```

#### sort()

***Customize Sort Function***

```python
def myfunc(n):
    return len(n)  # [2,5,3,6]

thislist = ['an', 'apple', 'the', 'banana']

# strings with small length to large length
thislist.sort(key = myfunc) # asc
print(thislist)
```

```python
def myfunc(n):
    return len(n)  # [2,5,3,6]

thislist = ['an', 'apple', 'the', 'banana']

# strings with large length to small length
thislist.sort(key = myfunc, reverse=True) # desc
print(thislist)
```

```python
def myfunc(n):
    return abs(n - 50)  # n-50: 50, 0, 15, 32, -27 => abs(n-50): 50, 0, 15, 32, 27 # index in asc: 1,2,4,3,0

thislist = [100, 50, 65, 82, 23]

thislist.sort(key = myfunc) # asc
print(thislist)
```

***Case Insensitive Sort***

```python
thislist = ["banana", "Orange", "Kiwi", "cherry"]
thislist.sort() # case sensitive sort
print(thislist)
```

```python
thislist = ["banana", "Orange", "Kiwi", "cherry"]

thislist.sort(key = str.lower, reverse=True) # case insensitive sort

print(thislist)

thislist.sort(key = str.upper, reverse=True) # case insensitive sort

print(thislist)
```

```python
def myFunc(e):
    return e['year'] # 2005, 2000, 2019, 2011

cars = [
  {'car': 'Ford', 'year': 2005},
  {'car': 'Mitsubishi', 'year': 2000},
  {'car': 'BMW', 'year': 2019},
  {'car': 'VW', 'year': 2011}
]

cars.sort(key=myFunc)

for x in cars:
  print(x)
```
