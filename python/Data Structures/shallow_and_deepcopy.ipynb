```python
a = [1,2,3,[4,5,6]]
# b = a # no copy 

# b = a[:] # shallow copy
# b = list(a) # shallow copy
# b = a.copy() # shallow copy

import copy
# b = copy.copy(a) # shallow copy
b = copy.deepcopy(a) # deep copy

# print(id(a),id(b))
# print(id(a[0]),id(b[0]))
# print(id(a[3][0]),id(b[3][0]))


# print(id(a) == id(b),
#       id(a[0]) == id(b[0]),
#       id(a[3][0]) == id(b[3][0]))

print('a:',a)
print('b:',b)
print("Updating b[0] and b[3][0]\n")
b[0] = 10
b[3][0] = 40

# print(id(a) == id(b),
#       id(a[0]) == id(b[0]),
#       id(a[3][0]) == id(b[3][0]))

print('a:',a)
print('b:',b)
```

**NOTE**
- Shallow copy: outer layer is copied. Nested layers are not copied.
    - achieved by slicing, constructor, copy() method, copy() function in copy module
- Deep copy: Recurssive copying is done (outer and all nested layers are copied)
    - achieved by deepcopy() function in copy module
