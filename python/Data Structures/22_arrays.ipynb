```python
# list
cars = ["Ford", "Volvo", "BMW"]

print(cars)
print(type(cars))
```

```python
# creating an array containing same
# data type elements
import array as arr

abc = arr.array('i', [1, 2, 3])

# accessing elements of array
for i in abc:
    print(i)

print(abc)
print(type(abc))
```
