### Tuple

- tuple is a collection which is ordered and unchangeable/immutable
- written with round brackets
- allow duplicate values
- tuples are indexed, the first item has index [0], the second item has index [1] etc.

#### create tuple

```python
thistuple = ("apple", "banana", "cherry")
print(thistuple)

thistuple = ()
print(thistuple, type(thistuple))
```

- limited inbuilt methods | Several inbuilt methods
- **historical, db data** | temp data

#### loop tuple items

```python
thistuple = ("apple", "banana", "cherry")
for x in thistuple:
  print(x)
```

```python
thistuple = ("apple", "banana", "cherry")
i = 0
while i < len(thistuple): # len(thistuple) = 3
  print(thistuple[i])
  i = i + 1
# .... out of loop line

# i=0
# 0<3 : True : print t[0] => 'apple' ; i=1
# 1<3 : True : print t[1] => 'banana' ; i=2
# 2<3 : True : print t[2] => 'cherry' ; i=3
# 3<3 : False => out of loop


i = 0
while i < 2:
  print(thistuple[i])
  i = i + 1
# .... out of loop line


# i=0
# 0<2 : True : print t[0] => 'apple' ; i=1
# 1<2 : True : print t[1] => 'banana' ; i=2
# 2<2 : False =>  out of loop

i = 0
while i < len(thistuple): # loop through entire tuple
  print(thistuple[i])
  i = i + 2 # loop through alternate elements
# .... out of loop line
```

#### named tuple

- Like dictionaries, they contain keys that are hashed to a particular value.
- But on contrary, it supports both access from key-value and iteration, the functionality that dictionaries lack.
- elements are ordered

```python
from collections import namedtuple

# Declaring namedtuple()
Student = namedtuple('Student', ['name', 'age', 'DOB']) # Student is tuple name. 'name', 'age', 'DOB' are keys of that tuple

# Adding values
S = Student('Nandini', '19', '25-4-1997')

# Access using index
print('Student age:',S[1])

# Access using name
print('Student name:',S.name)

# Access using getattr()
print('Student DOB:',getattr(S, 'DOB'))
```

#### conversion operations of named tuple

```python
S, type(S)
```

```python
# using _asdict() to return an OrderedDict()
print(S._asdict())

# initializing iterable
li = ['Manjeet', '19', '411997']

# using _make() to return namedtuple()
print(Student._make(li))

# initializing dict
di = {'name': "Nikhil", 'age': 19, 'DOB': '1391997'}

# using ** operator to return namedtuple from dictionary
print(Student(**di))

# using _fields to display all the keynames of namedtuple()
print(S._fields)

# ._replace returns a new namedtuple, it does not modify the original
print(S._replace(name='Manjeet'))
print(S)
```

```python
from collections import namedtuple

Point = namedtuple('Point', ['x', 'y'])
p1 = Point(x=1, y=2)
print(type(p1))
print(p1.x, p1.y) # Output: 1 2

p2 = Point(x=3, y=4)
print(type(p1))
print(p2.x, p2.y) # Output: 1 2
```
